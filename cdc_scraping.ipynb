```python
with open('cdc.txt', 'a', encoding='utf-8') as f:
    # Check if the file is empty and write the header if it is
    if f.tell() == 0:
        f.write('overview, symptoms, risk, prevention\n')
    
    overview = soup.find('div', attrs={'data-section': 'cdc_disease_basics_overview'})
    overview_text = overview.text.strip() if overview else 'N/A'
    
    symptoms = soup.find('div', attrs={'data-section': 'cdc_disease_basics_symptoms'})
    symptoms_text = symptoms.text.strip() if symptoms else 'N/A'
    
    risk = soup.find('div', attrs={'data-section': 'cdc_disease_basics_risk'})
    risk_text = risk.text.strip() if risk else 'N/A'
    
    prevention = soup.find('div', attrs={'data-section': 'cdc_disease_basics_prevention'})
    prevention_text = prevention.text.strip() if prevention else 'N/A'
    
    f.write(f'Overview: {overview_text}\nSymptoms: {symptoms_text}\nRisk: {risk_text}\nPrevention: {prevention_text}\n')
```
